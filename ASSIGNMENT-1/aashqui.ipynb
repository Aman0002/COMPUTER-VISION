```
#SOLUTION-1 
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # for image display
from skimage import io

#URL from where the image is downloaded
url = "/IMG-20201225-WA0207.jpg"

#Reading the image from the URL
myImg = io.imread(url)  

#Open-cv reads the image in BGR format so to change it to RGB
original_img = cv.cvtColor(myImg, cv.COLOR_BGR2RGB)

#Image is showing
print("\n\n")
print("**IIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIII MISSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSSS UHHHHHHHHHHHHHHHHHHHH**")
cv2_imshow(original_img)
print("\n")
#Getting the dimensions of our image height,width,number of channels R,G,B
dimensions = myImg.shape


#print(np.mean(original_img,axis=0))


for i in range (0,dimensions[0]):
  for j in range (0,dimensions[1]):
    if (original_img[i,j,0]==6 and original_img[i,j,1]==6 and original_img[i,j,2]==6 ):
      print("Index are x:{} , y:{}".format(i,j))

avg_0 = np.average(original_img[:, :, 0])

avg_1 = np.average(original_img[:, :, 1])

avg_2 = np.average(original_img[:, :, 2])

print("\n")
for i in range (0,dimensions[0]):
  for j in range (0,dimensions[1]):
        original_img[i,j,0]-=avg_0
        original_img[i,j,1]-=avg_1
        original_img[i,j,2]-=avg_2

cv2_imshow(original_img)
```
