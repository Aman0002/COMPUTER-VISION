```
#SOLUTION-1 
import numpy as np
import cv2 as cv 
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # for image display
from skimage import io

#URL from where the image is downloaded
url = "https://placekitten.com/500/300"

#Reading the image from the URL
myImg = io.imread(url)  

#Open-cv reads the image in BGR format so to change it to RGB
original_img = cv.cvtColor(myImg, cv.COLOR_BGR2RGB)

#Image is showing
cv2_imshow(original_img)
print("\n")
#Getting the dimensions of our image height,width,number of channels R,G,B
dimensions = myImg.shape
print("Image Dimensions:",dimensions)
print("Image height:",dimensions[0])
print("Image Width:",dimensions[1])
print("Image Numer of Channels:",dimensions[2])
```
