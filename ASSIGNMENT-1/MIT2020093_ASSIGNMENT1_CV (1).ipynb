## **Ques-2.** Perform following operations on an image:

**Ques-2.(a)** 

          Grayscale conversion.
          

```
#SOLUTION-2.(a)
print("\nSolution-2 (a)\n")
gray_img = cv.cvtColor(myImg,cv.COLOR_BGR2GRAY)#converting B-G-R to Gray Image
print("Grayscale Conversion Succesfull!\n")
cv2_imshow(gray_img)
```

**Ques-2.(b)** 

    Find the index of pixels having value equal to 6 in both images.
    

```
#(b)
img_shape = myImg.shape
row = img_shape[0]
col = img_shape[1]

print("\nSolution-2 (b)\n")
#IN GRAY IMAGE
print("Index of Pixels having value 6 in Gray Image are:\n")
for i in range(row):
  for j in range(col):
    if (gray_img[i,j]==6):
      print("x:{} y:{}".format(i,j))

#IN COLOURED IMAGE
print("\nIndex of Pixels having value 6 in Coloured Image are:\n")
for i in range(row):
  for j in range(col):
    if (myImg[i,j,0]==6 and myImg[i,j,1]==6 and myImg[i,j,2]==6):
      print("x:{} y:{}".format(i,j))
```

**Ques-2.(c)**   

    Identify and display the value of darkest pixel in both the images.
    

```
#Getting Darkest pixel in Gray_Image and Coloured_Image

dark = 255 

for i in range (row):
  for j in range (col):
    if (gray_img[i,j]<dark):
      dark = gray_img[i,j]
      x = i
      y = j
 


print("Index of the Darkest Pixel in Gray Image is x: {} y: {} value:{}".format(x,y,dark))


#print("Index of the Darkest Pixel in Coloured Image is x: {} y: {} value:{}".format(x1,y1,dark_clr))
dark_clr = myImg[x,y,0] + myImg[x,y,1] + myImg[x,y,2] 
dark_clr/=3
print("Index of the Darkest Pixel in Coloured Image is x: {} y: {} value:{}".format(x,y,dark_clr))

```

## **Ques-3.** Compute the average pixel value for each channel R, G, and B then subtract the average value per channel for the above image ( used in q1).

```
#SOLUTION-3

# myImg is the image used in question-1
original_img = myImg.copy()

cv2_imshow(original_img)

pixels = row * col
print("\nBefore Operation The image looks as above:\n")

#getting averages pixel value of each channels B-G-R

avg_0 = np.sum(original_img[:,:,0]) / pixels

avg_1 = np.sum(original_img[:,:,1]) /pixels

avg_2 = np.sum(original_img[:,:,2]) / pixels


for i in range (0,dimensions[0]):

  for j in range (0,dimensions[1]):

        original_img[i,j,0]-=avg_0
        original_img[i,j,1]-=avg_1
        original_img[i,j,2]-=avg_2

cv2_imshow(original_img)
print("\nAfter subtracting the average value per channel,Image looks like as above:\n")
print("Average Blue ",avg_0)
print("Average Green ",avg_1)
print("Average Red ",avg_2)
```

**Ques-4.(c)**
  
    Shifting (0,0) to (10, 10)

```
#Translation Matrix
M = np.float32([[1, 0, 10], [0, 1, 10]]) 


res = cv.warpAffine(myImg, M , (col, row)) 
  
print("\n Image After Shifting:\n")
cv2_imshow(res)
```
